```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the pickel file
with open('../tracker_stubs/player_detections.pkl', 'rb') as f:
    player_positions = pickle.load(f)
```

```python
player_positions
```

```python
rows = []

for frame_idx, frame in enumerate(player_positions):
    for player_id, box in frame.items():
        x1, y1, x2, y2 = box
        center_x = (x1 + x2) / 2
        center_y = (y1 + y2) / 2
        rows.append({
            "frame": frame_idx,
            "player_id": player_id,
            "x1": x1,
            "y1": y1,
            "x2": x2,
            "y2": y2,
            "center_x": center_x,
            "center_y": center_y
        })

df_players = pd.DataFrame(rows)
```

```python
df_players.to_pickle("dataframe/df_player_positions.pkl")
```
